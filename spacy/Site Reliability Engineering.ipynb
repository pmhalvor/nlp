```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url = 'https://sre.google/sre-book/introduction/'
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content, 'lxml')
page = soup.text.split('Bibliography')[1] # get rid of nav bar links
page = page.split('Previous')[0]          # get rid of footer
page = page.replace('\n\n\n\n', '')         # cut unnecessary whitespaces
# page = page.replace("’", "'")               # replace erroneous apostrophe
    
print(page)
```

```python
h1s = []

for h1 in soup.find_all('h1'):
    h1s.append(h1.text)
    page = page.replace(f'\n\n{h1.text}', f'\n\n# {h1.text}') 
```

```python
h2s = []
for h2 in soup.find_all('h2'):
    h2s.append(h2.text)
    page = page.replace(f'\n\n{h2.text}', f'\n\n## {h2.text}') 
```
